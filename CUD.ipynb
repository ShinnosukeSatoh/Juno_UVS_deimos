カラーユニバーサルデザイン(CUD)対応のカラーパレット

```python
# Matplotlibデフォルトの場合
import numpy as np
import matplotlib.pyplot as plt

x =  np.linspace(0,2*np.pi,100)
fig, ax = plt.subplots(figsize=(5,2.5),dpi=120)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
for i in range(6):
    a = np.random.rand()
    b = np.random.rand()
    ax.plot(x, a*np.sin(b*np.pi*x), label='#'+str(i))
ax.legend(ncol=3)
fig.tight_layout()
plt.show()
```
